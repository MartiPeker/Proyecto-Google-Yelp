```python
import pyarrow.parquet as pq
import os
import json
import pandas as pd
```

Reutilizo las funciones 'iterador_linea' e 'iterador_archivo' que creé previamente en el notebook de filter-states-year.ipynb para unir todos los json existentes en data en un nuevo dataframe y convertirlo a .parquet para analizar la totalidad de los datos allí.

```python
def iterador_linea(archivo_json):
    objetos_json = []
    with open(archivo_json, "r") as file:
        for linea in file:
            objeto = json.loads(linea)
            objetos_json.append(objeto)
            
    df = pd.DataFrame(objetos_json)
    return df
```

```python
def iterador_archivo(carpeta):
    dataframes = []
    for archivo in os.listdir(carpeta):
        ruta_archivo = os.path.join(carpeta, archivo)
        if os.path.isfile(ruta_archivo):
            dataframes.append(iterador_linea(ruta_archivo))
    df = pd.concat(dataframes, axis=0, ignore_index=True)
    return df
```

```python
metadata = "../metadata-sitios/"
df = iterador_archivo(metadata)
```

Para trabajar con la columna 'category' y buscar todos aquellos comercios que contengan la palabra 'lodging' debo transformar los objetos/arrays en str.

```python
def convert_str(arr):
    if arr is None:
        return ""
    elif isinstance(arr, list):
        return ', '.join(map(str, arr))
    else:
        return str(arr)
```

Convierto los arrays de la columna en str:

```python
df['category'] = df['category'].apply(convert_str)
```

```python
palabras = 'lodging'

hoteles = df[df['category'].str.contains(palabra)]
```
