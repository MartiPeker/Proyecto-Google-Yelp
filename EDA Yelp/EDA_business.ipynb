```python
import pandas as pd
```

```python
df = pd.read_pickle("business.pkl")
```

```python
df.info()
```

Eliminar columnas duplicadas:

```python
df = df.loc[:, ~df.columns.duplicated()]
```

```python
df.info()
```

```python
df.head(5)
```

Alineamos las primeras 3 filas de la columna "state":

```python
# Número de filas a desplazar hacia arriba
num_rows = 3

# Desplazar la columna hacia arriba
df["state"] = df["state"].shift(-num_rows)
```

```python
df.head()
```

Duplicados en la columna "categories":

```python
duplicates_count = df['categories'].duplicated().sum()

print("Cantidad de duplicados en 'categories':", duplicates_count)
```

```python
df = df.dropna(subset=['categories'])
```

Filtro los business_id en los que la categoría empieza con "hotel":

```python
df1= df[df['categories'].str.lower().str.startswith('hotel')]
```

```python
df1.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

10 estados que contienen más hoteles:

```python
# Calcular la frecuencia de cada estado
state_counts = df1["state"].value_counts()

# Tomar el top 12 estados más comunes
top_states = state_counts.head(10)

green_palette = sns.color_palette("Greens_r", len(top_states))

# Crear un gráfico de barras para visualizar los estados más comunes
plt.figure(figsize=(10, 6))
top_states.plot(kind='bar', color=green_palette)
plt.xlabel('Estado')
plt.ylabel('Cantidad Hoteles')
plt.title('Top 10 Estados')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

10 ciudades con más hoteles:

```python
city_counts = df1["city"].value_counts()

top_cities = city_counts.head(10)

plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar', color=green_palette)
plt.xlabel('Ciudades')
plt.ylabel('Cantidad Hoteles')
plt.title('Top 10 Ciudades')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
